# EDA Données VLille

```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_parquet("data_vlille_2025.parquet")
```

```python
df["ID_station"] = df["ID_station"].str.replace("vlille_temps_reel.", "", regex=False)
df["Date_Paris"] = pd.to_datetime(df["properties_date_modification"]).dt.tz_convert("Europe/Paris") 
df["Date_Scrapping"]= df["Date_Scrapping"].str.replace("_", "/", regex=False)
df["Date_Scrapping"] = pd.to_datetime(df["Date_Scrapping"], format = "%d/%m/%Y")
```

```python
import plotly.express as px
from datetime import date
df = df.sort_values("Date_Paris")
df = df[df["Date_Scrapping"] >= pd.Timestamp("2025-01-01")]
to_plot = df.groupby(df["Date_Paris"].dt.floor("h"))["properties_nb_places_dispo"].mean().round()
px.line(to_plot, x=to_plot.index, y="properties_nb_places_dispo", title="Moyenne du nombre de places disponibles au fil du temps")
# date_ref = pd.Timestamp("2025-11-14", tz="Europe/Paris")
# df_angellier = df[df["properties_nom"] == "ANGELLIER"]
# df_angellier = df_angellier[df_angellier["Date_Paris"] >= date_ref]
# df_angellier = df_angellier.sort_values("Date_Paris")
# px.line(df_angellier, x=df_angellier.Date_Paris, y="properties_nb_places_dispo", title="Moyenne du nombre de places disponibles au fil du temps")
```
