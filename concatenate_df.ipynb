```python
import pandas as pd
import os
```

```python
path = os.getcwd()
path = path +'/data/'
i = 0

month_dic = {'01':'January','02':'February','03':'March',
             '04':'April','05':'May','06':'June',''
             '07':'July','08':'August','09':'September',
             '10':'October','11':'November','12':'December'}

month_dic_file = { "01" : [], "02" : [], "03" : [],
                   "04" : [], "05" : [], "06" : [],
                    "07" : [], "08" : [], "09" : [],
                    "10" : [], "11" : [], "12" : [] }

for file in os.listdir(path) :
    if file.endswith(".csv") : 
        name_file = file.split('.')[0]
        month_dic_file[name_file[15:17]].append(file)

for month in month_dic_file :
    joined_df = pd.DataFrame()
    for file in month_dic_file[month] :
        print(file)
        df = pd.read_csv(path+file)
        df["Scrapped_Date"] = file[12:22]
        df["Scrapped_Hour"] = file[23:30]

        joined_df = pd.concat([joined_df,df], ignore_index=True)
    joined_df.to_csv(path+"Joined_DF_"+month_dic[month]+'.csv', mode='a', header=False, index=False)


print(month_dic_file)

    
        
    
```

```python
month_to_delete = ["03", "04", "05", "06", "07", "08", "09", "10"]
delete_file_dic = {}
for file in os.listdir(path) :
    if file.endswith(".csv") : 
        if file.startswith("vlille_") :
            name_file = file.split('.')[0]
            month_of_file = name_file[15:17]
            if month_of_file in month_to_delete:
                os.remove(path+file)
```
